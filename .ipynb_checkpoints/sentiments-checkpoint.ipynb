```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import razdel
import re
import csv
from nltk import TweetTokenizer
import transformers
import torch
import fasttext
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, roc_auc_score, f1_score
from sklearn.model_selection import train_test_split
```

### Сбор данных

```python
neg_corus = pd.read_csv('data/negative.csv', sep=";")
pos_corus = pd.read_csv('data/positive.csv', sep=";")

neg_corus.columns = ['id', 'tdate', 'tmane', 'ttext', 'ttype', 'trep', 'trtw','tfav', 'tstcount', 'tfol', 'tfrien', 'listcount']
pos_corus.columns = ['id', 'tdate', 'tmane', 'ttext', 'ttype', 'trep', 'trtw','tfav', 'tstcount', 'tfol', 'tfrien', 'listcount']
```

```python
def load_data(filename):
    tokenizer = TweetTokenizer()

    with open(filename, 'r') as f:
        reader = csv.DictReader(f)
        samples = []
        labels = []
        for row in reader:
            text = row['text']
            label = row['label']

            text_tokenized = tokenizer.tokenize(text)
            text_joined = ' '.join(text_tokenized)

            samples.append(text_joined)
            labels.append(label)

    return samples, labels

tokenizer = razdel
def filter_text(text):
    line = tokenizer.tokenize(text.lower())
    line = [token.text for token in line]
    filtered_line = ' '.join([token for token in line if token.isalpha()])
        
    return filtered_line
```

```python
samples_rusent_train, labels_rusent_train = load_data('data/rusentiment_preselected.csv')
samples_rusent_test, labels_rusent_test = load_data('data/rusentiment_test.csv')

samples_pos_train = np.array(samples_rusent_train)[np.array(labels_rusent_train) == 'positive'].tolist()
samples_neg_train = np.array(samples_rusent_train)[np.array(labels_rusent_train) == 'negative'].tolist()
samples_pos_test = np.array(samples_rusent_test)[np.array(labels_rusent_test) == 'positive'].tolist()
samples_neg_test = np.array(samples_rusent_test)[np.array(labels_rusent_test) == 'negative'].tolist()
```

```python
train_common_list = []
test_common_list = []
train_rusent_list = []
test_rusent_list = []

for text in neg_corus['ttext']:
    train_common_list.append({"text":filter_text(text), "label":0})
for text in pos_corus['ttext']:
    train_common_list.append({"text":filter_text(text), "label":1})

for text in samples_neg_train:
    train_common_list.append({"text":filter_text(text), "label":0})
    train_rusent_list.append({"text":filter_text(text), "label":0})
for text in samples_pos_train:
    train_common_list.append({"text":filter_text(text), "label":1})
    train_rusent_list.append({"text":filter_text(text), "label":1})

for text in samples_neg_test:
    test_rusent_list.append({"text":filter_text(text), "label":0})
for text in samples_pos_test:
    test_rusent_list.append({"text":filter_text(text), "label":1})
```

```python
# _rusent_list - тексты из датасета RuSentiment
# train_common_list - тексты из датасетов RuSentiment и Corus 

len(train_common_list), len(train_rusent_list), len(test_rusent_list)
```

### Датасет и даталоадеры

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
bert_tokenizer = transformers.BertTokenizer.from_pretrained('DeepPavlov/rubert-base-cased')
```

```python
class SentimentDataset(torch.utils.data.Dataset):
    def __init__(self, dataset, tokenizer=bert_tokenizer, max_length=16):
        self.dataset = dataset
        self.tokenizer = tokenizer
        self.max_length = max_length
        
    def __getitem__(self, idx):
        text = self.dataset[idx]['text']
        label = self.dataset[idx]['label']
        
        tokenizer_output = self.tokenizer.encode_plus(
            text,
            max_length = self.max_length,
            return_tensors = 'pt',
            padding = 'max_length')
        
        return {
            "input_ids": tokenizer_output['input_ids'].squeeze(0)[:self.max_length], 
            "mask": tokenizer_output['attention_mask'].squeeze(0)[:self.max_length],
            "label": label
        }
        
    def __len__(self):
        return len(self.dataset)
```

```python
# downsampling для преобладающего класса
def balance(train_list):
    neg_labels = sum(np.array([text_dict['label'] for text_dict in train_list]) == 0)
    pos_labels = sum(np.array([text_dict['label'] for text_dict in train_list]) == 1)
    diff = pos_labels - neg_labels

    counter = 0
    for sample in train_list:
      if sample['label'] == 1 and counter<diff:
        train_list.remove(sample)
        counter+=1
      if counter == diff:
        break
      
    return train_list
```

```python
train_common_list = balance(train_common_list)
train_rusent_list = balance(train_rusent_list)
```

```python
train_common_list_, test_common_list_ = train_test_split(train_common_list, train_size = 0.8, 
                                            stratify = [text_dict['label'] for text_dict in train_common_list])
```

```python
train_common_dataset = SentimentDataset(train_common_list_)
test_common_dataset = SentimentDataset(test_common_list_)

train_rusent_dataset = SentimentDataset(train_rusent_list)
test_rusent_dataset = SentimentDataset(test_rusent_list)
```

```python
batch_size = 800

train_common_loader = torch.utils.data.DataLoader(
    train_common_dataset,
    batch_size = batch_size,
    shuffle = True
)

test_common_loader = torch.utils.data.DataLoader(
    test_common_dataset,
    batch_size = batch_size)

train_rusent_loader = torch.utils.data.DataLoader(
    train_rusent_dataset,
    batch_size = batch_size,
    shuffle = True
)

test_rusent_loader = torch.utils.data.DataLoader(
    test_rusent_dataset,
    batch_size = batch_size)
```

### Функции для обучения, теста, подсчета метрик

```python
def training(model, train_loader, valid_loader, weights_file, num_epochs=3):
    
    best_valid_acc = 0
    model.train()
    for e in range(num_epochs):
        model.train()
        train_loss = 0
        for batch in tqdm(train_loader):
            tokens = batch['input_ids'].to(device)
            mask = batch['mask'].to(device)
            target = batch['label'].to(device)
            
            optimizer.zero_grad()
            outputs = model(input_ids = tokens, attention_mask = mask, labels = target)
            loss = criterion(outputs.logits.squeeze(1), target) 
            loss.backward()
            optimizer.step()
            train_loss+= loss.item()
            
        valid_loss = 0
        valid_acc = 0
        model.eval()
        with torch.no_grad():
            for batch in valid_loader:
                tokens = batch['input_ids'].to(device)
                mask = batch['mask'].to(device)
                target = batch['label'].to(device)

                outputs = model(input_ids = tokens, attention_mask = mask, labels = target)
                loss = criterion(outputs.logits.squeeze(1), target)
              
                valid_loss += loss.item()
                valid_acc += sum(torch.argmax(outputs.logits, dim=-1)==target)/len(target)

        print(f"Train Loss: {train_loss / len(train_loader)} ,"
              f"Valid Loss: {valid_loss / len(valid_loader)},"
              f"Valid Acc: {valid_acc / len(valid_loader)}")
        
        if valid_acc > best_valid_acc:
            best_valid_acc = valid_acc
            torch.save(model.state_dict(), weights_file)
        
def testing(loader, model):
    y_label = []
    y_pred = []
    test_acc = 0
    model.eval()
    with torch.no_grad():
        for batch in tqdm(loader):
            tokens = batch['input_ids'].to(device)
            mask = batch['mask'].to(device)
            target = batch['label'].to(device)


            outputs = model(input_ids = tokens, attention_mask = mask, labels = target)
            test_acc += sum(torch.argmax(outputs.logits, dim=-1)==target)/len(target)

            y_label.extend(batch['label'])
            y_pred.extend(outputs.logits.cpu().numpy().tolist())

    print(f"Test Accuracy: {test_acc / len(loader)}")
    return y_label, y_pred

def calc_roc(y_pred, y_label, modelname):
    f1 = f1_score(np.argmax(np.array(y_pred), axis=-1), y_label)
    print("F1 score: ", f1)

    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(2):
        fpr[i], tpr[i], _ = roc_curve(np.array(np.array(y_label) == i), np.array(y_pred)[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    y_true = np.array(list(zip(np.array(np.array(y_label) == 0), np.array(np.array(y_label) == 1))))
    # Compute micro-average ROC curve and ROC area
    fpr["micro"], tpr["micro"], _ = roc_curve(y_true.ravel(), np.array(y_pred).ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    plt.figure()
    lw=2
    plt.plot(fpr[0], tpr[0], color='darkorange',
            lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[0])
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC for ' + modelname)
    plt.legend(loc="lower right")
    plt.show()
```

### FastText

```python
def create_wordlist(text):
    line = tokenizer.tokenize(text.lower())
    line = [token.text for token in line]
    return ' '.join(line)
```

```python
train_common_texts = [create_wordlist(text_dict["text"]) for text_dict in train_common_list_]
train_common_labels = [text_dict["label"] for text_dict in train_common_list_]

train_rusent_texts = [create_wordlist(text_dict["text"]) for text_dict in train_rusent_list]
train_rusent_labels = [text_dict["label"] for text_dict in train_rusent_list]

test_common_texts = [create_wordlist(text_dict["text"]) for text_dict in test_common_list_]
test_common_labels = [text_dict["label"] for text_dict in test_common_list_]

test_rusent_texts = [create_wordlist(text_dict["text"]) for text_dict in test_rusent_list]
test_rusent_labels = [text_dict["label"] for text_dict in test_rusent_list]
```

```python
df_train_common = pd.DataFrame({'text': train_common_texts, 'target': train_common_labels})
df_train_rusent = pd.DataFrame({'text': train_rusent_texts, 'target': train_rusent_labels})

df_test_common = pd.DataFrame({'text': train_common_texts, 'target': train_common_labels})
df_test_rusent = pd.DataFrame({'text': test_rusent_texts, 'target': test_rusent_labels})
```

```python
df_train_common['target'] = df_train_common['target'].apply(lambda x: '__label__' + str(x))
df_test_common['target'] = df_test_common['target'].apply(lambda x: '__label__' + str(x))

df_train_rusent['target'] = df_train_rusent['target'].apply(lambda x: '__label__' + str(x))
df_test_rusent['target'] = df_test_rusent['target'].apply(lambda x: '__label__' + str(x))

#making files for fasttext training
df_train_common.to_csv('df_train_common.txt', header=False, index=False, sep="\t")
df_test_common.to_csv('df_test_common.txt', header=False, index=False, sep="\t")

df_train_rusent.to_csv('df_train_rusent.txt', header=False, index=False, sep="\t")
df_test_rusent.to_csv('df_test_rusent.txt', header=False, index=False, sep="\t")
```

```python
def predict_sample(text, model):
    prediction = model.predict(text, k=1)
    tag = prediction[0][0]
    proba = prediction[1][0]
    tag = tag.replace('__label__', '')
    return int(tag), proba

def predict(test_list, model):
  y_pred = []
  y_label = []
  
  for text_dict in test_list:
    pred_label, proba = predict_sample(text_dict['text'], model)
    
    if pred_label==1:
      y_pred.append([1-proba, proba])
    else:
      y_pred.append([proba, 1-proba])

    y_label.append(text_dict['label'])
  
  return np.array(y_pred), np.array(y_label)
```

### Для RuSentiment + Corus датасетов

```python
model_common = fasttext.train_supervised(input='df_train_common.txt', dim=300,
                                  epoch=50, lr=0.1,
                                  wordNgrams=3, ws=5)

model_common.test("df_test_common.txt", k=1)
```

```python
y_pred, y_label = predict(test_common_list_, model_common)
calc_roc(y_pred, y_label, 'FastText')
```
